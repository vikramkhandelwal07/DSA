```python
class Solution:
    def findTarget(self, root, k):
        seen = set()
        
        def dfs(node):
            if not node:
                return False
            if (k - node.val) in seen:
                return True
            seen.add(node.val)
            return dfs(node.left) or dfs(node.right)
        
        return dfs(root)
```

```python
class BSTiterator:
    def __init__(self, root, forward=True):
        self.stack = []
        self.forward = forward
        self._push(root)
        
    def _push(self, node):
        while node:
            self.stack.append(node)
            node = node.left if self.forward else node.right
    
    def next(self):
        node = self.stack.pop()
        if self.forward:
            self._push(node.right)
        else:
            self._push(node.left)
        return node.val
    
    def hasnext(self):
        return bool(self.stack)


def findTarget(root, target):
    if not root:
        return False

    leftItr = BSTiterator(root, True)    # smallest → largest
    rightItr = BSTiterator(root, False)  # largest → smallest

    if not leftItr.hasnext() or not rightItr.hasnext():
        return False

    leftVal = leftItr.next()
    rightVal = rightItr.next()
    
    while leftVal < rightVal:  # ensure we don't use the same node twice
        s = leftVal + rightVal
        if s == target:
            return True
        elif s < target:
            if leftItr.hasnext():
                leftVal = leftItr.next()
            else:
                break
        else:
            if rightItr.hasnext():
                rightVal = rightItr.next()
            else:
                break
    return False
    
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

class Solution:
    def findTarget(self, root: Optional[TreeNode], k: int) -> bool:
        if not root:
            return False

        # Morris forward iterator: yields next smallest
        def next_smallest():
            curr = root
            while curr:
                if not curr.left:
                    val = curr.val
                    curr = curr.right
                    yield val
                else:
                    pred = curr.left
                    while pred.right and pred.right is not curr:
                        pred = pred.right
                    if not pred.right:
                        pred.right = curr
                        curr = curr.left
                    else:
                        pred.right = None
                        val = curr.val
                        curr = curr.right
                        yield val

        # Morris reverse iterator: yields next largest
        def next_largest():
            curr = root
            while curr:
                if not curr.right:
                    val = curr.val
                    curr = curr.left
                    yield val
                else:
                    succ = curr.right
                    while succ.left and succ.left is not curr:
                        succ = succ.left
                    if not succ.left:
                        succ.left = curr
                        curr = curr.right
                    else:
                        succ.left = None
                        val = curr.val
                        curr = curr.left
                        yield val

        it_l = next_smallest()
        it_r = next_largest()

        try:
            left = next(it_l)
            right = next(it_r)
        except StopIteration:
            return False

        # advance until pointers cross (must ensure they don't point to same node value twice)
        while left < right:
            s = left + right
            if s == k:
                return True
            if s < k:
                try:
                    left = next(it_l)
                except StopIteration:
                    break
            else:
                try:
                    right = next(it_r)
                except StopIteration:
                    break

        return False
```
